```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy #funcoes de treinamento
from scipy.optimize import curve_fit # otmizaçao MSE
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

```

```python
data = pd.read_csv("headbrain.csv")
data.head()
```

```python
#converter em array

x = data["Head Size(cm^3)"].values # a variavel que usa para prever
y = data["Brain Weight(grams)"].values # variavel que estamos querendo prever
```

```python
#Modelo de regressão

reg = LinearRegression()
x_Reshaped=x.reshape((-1,1))# formato 2d
regressao = reg.fit (x_Reshaped,y)  # encontra os coef (realiza a regressao)
```

# Nova seção

```python
previsao = reg.predict(x_Reshaped) #realiza a previsao
```

```python
print('Y = {}  X= {}'.format(reg.coef_,reg.intercept_))

R_2 = r2_score(y, previsao) #calculo R2
print("coeficiente de determinaçao (R2)", R_2)
```

```python
plt.figure(figsize=(4, 4), dpi=100)
plt.scatter(x, y, color='gray')  # plot de dispersão
plt.plot(x, previsao, color='red', linewidth=2)
plt.xlabel("Head Size(cm^3)")
plt.ylabel("Brain Weight(grams)")
plt.show()
```
